```python
all_kf_preds = pd.DataFrame()
all_kf_preds.index = df.index

for param_idx, params in enumerate(param_ls):
    param_key = f'params_{param_idx}'
    all_results[param_key] = {
        'parameters': params,
        'folds': {},
        'all_fold_preds': None
    }

    estimator = clone(fit.estimator).set_params(**params)

    for fold_idx, (train_idx, test_idx) in enumerate(cv_splits):
        kf_feat_tr = df.iloc[train_idx] if hasattr(df, 'iloc') else df[train_idx]
        kf_feat_te = df.iloc[test_idx] if hasattr(df, 'iloc') else df[test_idx]
        kf_targ_tr = targs.iloc[train_idx] if hasattr(targs, 'iloc') else targs[train_idx]
        kf_targ_te = targs.iloc[test_idx] if hasattr(targs, 'iloc') else targs[test_idx]

        all_results[param_key]['folds'][f'fold_{fold_idx}'] = {
            "kf_feat_tr": kf_feat_tr,
            "kf_feat_te": kf_feat_te,
            "kf_targ_tr": kf_targ_tr,
            "kf_targ_te": kf_targ_te
        }

        estimator.fit(kf_feat_tr, kf_targ_tr)

        test_preds = estimator.predict(kf_feat_te)
        fold_df = pd.DataFrame()
        fold_df['ID'] = kf_feat_te.index
        fold_df[f'preds'] = test_preds
        fold_df.set_index('ID')

        all_fold_preds = all_results[param_key]['all_fold_preds']
        all_fold_preds = pd.concat([all_fold_preds, fold_df], ignore_index=True)
        all_results[param_key]['all_fold_preds'] = all_fold_preds


    pred_ls = all_results[param_key]['all_fold_preds']['preds'].tolist()
    all_kf_preds[f'Param_set_{param_idx}'] = pred_ls

```

```python
import numpy as np

stats = pd.DataFrame(index=all_kf_preds.index)

stats['Mean_Pred'] = all_kf_preds.mean(axis=1)
stats['Docking_Score'] = targs['Affinity(kcal/mol)']
stats['Stdev_Pred'] = all_kf_preds.std(axis=1)
stats['Prediction_Error'] = stats['Mean_Pred'].astype(float) - targs['Affinity(kcal/mol)'].astype(float)
stats['Absolute_Prediction_Error'] = stats['Prediction_Error'].abs()
max_error = np.max(stats['Absolute_Prediction_Error'])
stats['Normalised_Error'] = stats['Absolute_Prediction_Error']/max_error if max_error > 0 else stats['Absolute_Prediction_Error']
stats['Reliability_Score'] = 1 - (stats['Normalised_Error']* stats['Stdev_Pred'])
```

```python
preds = model.predict(held_out_feat_df)
results_df = pd.DataFrame({
    'True_Affinity': held_out_targ_df,
    "ID": ids,
    'Predicted_Affinity': preds
})
results_df = results_df.set_index('ID')
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import pearsonr
import json

true = np.array(results_df['True_Affinity']).astype(float)
pred = np.array(results_df['Predicted_Affinity']).astype(float)

errors = true - pred
# Calculate performance metrics
bias = np.mean(errors)
sdep = (np.mean((true - pred - (np.mean(true - pred))) ** 2)) ** 0.5
mse = mean_squared_error(true, pred)
rmse = np.sqrt(mse)
r2 = r2_score(true, pred)
r_pearson, p_pearson = pearsonr(true, pred)

stats = {
    "Bias": round(bias, 3),
    "SDEP": round(sdep, 3),
    "MSE": round(mse, 3),
    "RMSE": round(rmse, 3),
    "r2": round(r2, 3),
    "pearson_r": round(r_pearson, 3),
    "pearson_p": round(p_pearson, 3),
}
stats
with open(f"{working_dir}/held_out_test/held_out_stats.json", "w") as f:
    json.dump(stats, f, indent=4)
```
